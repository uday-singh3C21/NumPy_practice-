```python
x_sin=np.arange(0,3*np.pi,0.1)
```

```python
print(x_sin)
```

```python
y_sin=np.sin(x_sin)
print(y_sin)
```

```python
plt.plot(x_sin,y_sin)
plt.show()
```

```python
y_cos=np.cos(x_sin)
plt.plot(x_sin,y_cos)
plt.show()
```

```python
y_tan=np.tan(x_sin)
plt.plot(x_sin,y_tan)
plt.show()
```
